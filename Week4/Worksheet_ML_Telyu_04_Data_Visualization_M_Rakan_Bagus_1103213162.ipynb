Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python

# Memilih kolom bertipe kategorikal
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Menampilkan pie chart untuk setiap kolom kategorikal
for column in categorical_columns:
    # Menghitung persentase setiap kategori
    value_counts = df[column].value_counts(normalize=True) * 100

    # Membuat pie chart
    plt.figure(figsize=(8, 6))
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Persentase Kategori untuk {column}')
    plt.show()
````

````python
def plot_pie_chart(column_name):
    # Mengambil data dari kolom yang diberikan
    data = df[column_name].value_counts()
    labels = data.index
    sizes = data.values

    # Memilih colormap yang estetis
    cmap = plt.get_cmap('Pastel1')  # Menggunakan colormap 'Pastel1' untuk tampilan lebih lembut
    colors = cmap.colors[:len(labels)]  # Mengambil warna sebanyak jumlah label

    # Membuat pie chart
    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'edgecolor': 'black'})
    plt.axis('equal')  # Untuk memastikan pie chart berbentuk lingkaran
    plt.title(f'Percentage of {column_name}', fontsize=14)
    plt.show()

# Memanggil fungsi untuk setiap kolom kategorikal
plot_pie_chart('gender')
plot_pie_chart('race/ethnicity')
plot_pie_chart('lunch')
plot_pie_chart('parental level of education')
plot_pie_chart('test preparation course')
````

````python
import matplotlib.pyplot as plt

def plot_score_distribution():
    # Mengatur ukuran figure
    plt.figure(figsize=(18, 5))

    # Histogram untuk 'math score'
    plt.subplot(1, 3, 1)
    plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Distribusi Math Score')
    plt.xlabel('Score')
    plt.ylabel('Jumlah Siswa')

    # Histogram untuk 'reading score'
    plt.subplot(1, 3, 2)
    plt.hist(df['reading score'], bins=20, color='lightgreen', edgecolor='black')
    plt.title('Distribusi Reading Score')
    plt.xlabel('Score')
    plt.ylabel('Jumlah Siswa')

    # Histogram untuk 'writing score'
    plt.subplot(1, 3, 3)
    plt.hist(df['writing score'], bins=20, color='lightcoral', edgecolor='black')
    plt.title('Distribusi Writing Score')
    plt.xlabel('Score')
    plt.ylabel('Jumlah Siswa')

    # Menampilkan plot
    plt.tight_layout()
    plt.show()

# Memanggil fungsi untuk menampilkan distribusi nilai
plot_score_distribution()
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python
import pandas as pd

def count_pass_fail(subject):
    # Mengkategorikan siswa berdasarkan nilai minimal 75
    pass_count = (df[subject] >= 75).sum()
    fail_count = (df[subject] < 75).sum()

    return pass_count, fail_count

# Menghitung jumlah pass dan fail untuk masing-masing pelajaran
math_pass, math_fail = count_pass_fail('math score')
reading_pass, reading_fail = count_pass_fail('reading score')
writing_pass, writing_fail = count_pass_fail('writing score')

# Menampilkan hasil
print(f"Math: Pass = {math_pass}, Fail = {math_fail}")
print(f"Reading: Pass = {reading_pass}, Fail = {reading_fail}")
print(f"Writing: Pass = {writing_pass}, Fail = {writing_fail}")
````

##INSIGHT

Math:
Jumlah siswa yang "pass" di matematika cenderung lebih rendah dibandingkan dengan "fail," mengindikasikan bahwa matematika adalah salah satu mata pelajaran yang lebih menantang bagi banyak siswa.
Banyak siswa mendapatkan nilai di bawah 75, yang menunjukkan bahwa ada kebutuhan untuk perbaikan dan dukungan tambahan dalam pengajaran matematika.

Reading:
Sebagian besar siswa berhasil mencapai nilai "pass" di reading, menunjukkan bahwa keterampilan membaca cenderung lebih kuat di antara siswa.
Jumlah siswa yang "fail" lebih sedikit dibandingkan dengan matematika, yang dapat menunjukkan bahwa pendekatan pengajaran dalam membaca lebih efektif atau siswa lebih mampu dalam pemahaman bacaan.

Writing:
Pola pada writing mirip dengan reading, di mana lebih banyak siswa yang "pass" dibandingkan yang "fail." Ini menunjukkan bahwa sebagian besar siswa memiliki keterampilan menulis yang cukup baik.
Meskipun demikian, jumlah siswa yang "fail" di writing sedikit lebih tinggi dibandingkan dengan reading, yang berarti ada ruang untuk perbaikan, terutama bagi siswa yang memiliki nilai di bawah 75.

Secara keseluruhan, hasil ini menunjukkan bahwa matematika cenderung menjadi tantangan yang lebih besar bagi siswa, sementara kemampuan reading dan writing secara umum lebih baik. Ini mungkin mengindikasikan perlunya perhatian lebih dalam mendukung siswa dalam matematika untuk meningkatkan tingkat kelulusan mereka.

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_score_by_category(category):
    # Mengatur ukuran figure
    plt.figure(figsize=(18, 12))

    # Boxplot untuk 'math score'
    plt.subplot(3, 1, 1)
    sns.boxplot(x=df[category], y=df['math score'], palette='Set2')
    plt.title(f'Math Score by {category}', fontsize=14)
    plt.xlabel(category)
    plt.ylabel('Math Score')

    # Boxplot untuk 'reading score'
    plt.subplot(3, 1, 2)
    sns.boxplot(x=df[category], y=df['reading score'], palette='Set3')
    plt.title(f'Reading Score by {category}', fontsize=14)
    plt.xlabel(category)
    plt.ylabel('Reading Score')

    # Boxplot untuk 'writing score'
    plt.subplot(3, 1, 3)
    sns.boxplot(x=df[category], y=df['writing score'], palette='Set1')
    plt.title(f'Writing Score by {category}', fontsize=14)
    plt.xlabel(category)
    plt.ylabel('Writing Score')

    # Menampilkan plot
    plt.tight_layout()
    plt.show()

# Visualisasi untuk masing-masing kategori
plot_score_by_category('gender')
plot_score_by_category('race/ethnicity')
plot_score_by_category('parental level of education')
plot_score_by_category('lunch')
plot_score_by_category('test preparation course')
````

Lunch: Perbedaan skor antara siswa yang mendapat makan siang gratis atau bersubsidi dan yang tidak dapat menunjukkan dampak sosial ekonomi terhadap prestasi akademis.

Test Preparation Course: Siswa yang mengikuti kursus persiapan tes mungkin memiliki distribusi nilai yang lebih tinggi, menunjukkan efektivitas persiapan.

Visualisasi ini memberikan gambaran yang lebih dalam tentang faktor-faktor yang mungkin memengaruhi kinerja akademis siswa.








Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Menghitung nilai rata-rata seluruh mata pelajaran
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

def plot_average_score_by_category(category):
    # Mengatur ukuran figure
    plt.figure(figsize=(10, 6))

    # Boxplot untuk nilai rata-rata berdasarkan kategori
    sns.boxplot(x=df[category], y=df['average_score'], palette='pastel')
    plt.title(f'Average Score by {category}', fontsize=14)
    plt.xlabel(category)
    plt.ylabel('Average Score')

    # Menampilkan plot
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Visualisasi untuk masing-masing kategori
plot_average_score_by_category('gender')
plot_average_score_by_category('race/ethnicity')
plot_average_score_by_category('parental level of education')
plot_average_score_by_category('lunch')
plot_average_score_by_category('test preparation course')
````

##INSIGHT

Gender:

Parental Level of Education:
